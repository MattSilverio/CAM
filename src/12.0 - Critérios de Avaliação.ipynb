```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import numpy as np

# Carregando o conjunto de dados Iris
iris = load_iris()
X, y = iris.data, iris.target

# Criando o modelo de Floresta Aleatória
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Obtendo a importância das características
feature_importances = rf.feature_importances_

# Plotando a importância das características
plt.barh(range(X.shape[1]), feature_importances, align='center')
plt.yticks(np.arange(X.shape[1]), iris.feature_names)
plt.xlabel('Importância da Característica')
plt.ylabel('Característica')
plt.title('Importância das Características com Floresta Aleatória')
plt.show()
```
